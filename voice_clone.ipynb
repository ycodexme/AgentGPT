```python
# Configuration de l'encodage
import sys
import locale
sys.stdout.reconfigure(encoding='utf-8')
locale.getpreferredencoding = lambda: 'UTF-8'

import os
import torch
from TTS.api import TTS
from IPython.display import Audio, display
from google.colab import files

class VoiceCloner:
    def __init__(self):
        """Initialise le cloneur de voix"""
        self.device = "cuda" if torch.cuda.is_available() else "cpu"
        print(f"Utilisation de : {self.device}")

        try:
            print("Chargement du modele...")
            self.tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(self.device)
            print("Modele charge avec succes!")
        except Exception as e:
            print(f"Erreur lors du chargement du modele : {str(e)}")
            raise

    def ensure_wav_extension(self, filename):
        """S'assure que le fichier a une extension .wav"""
        name, ext = os.path.splitext(filename)
        if not ext:
            return f"{name}.wav"
        elif ext.lower() != '.wav':
            return f"{name}.wav"
        return filename

    def clone_voice(self, text, reference_audio, output_path="output.wav", language="fr"):
        """Clone une voix"""
        try:
            # Assurer l'extension WAV
            output_path = self.ensure_wav_extension(output_path)

            print(f"Generation de l'audio pour le texte : {text}")
            self.tts.tts_to_file(
                text=text,
                speaker_wav=reference_audio,
                language=language,
                file_path=output_path
            )

            print(f"Audio genere avec succes : {output_path}")
            return Audio(output_path)

        except Exception as e:
            print(f"Erreur lors du clonage : {str(e)}")
            return None

def run_voice_cloner():
    print("Initialisation du cloneur de voix...")
    cloner = VoiceCloner()

    print("\nUploadez votre fichier audio de reference (format WAV)...")
    uploaded = files.upload()

    if not uploaded:
        print("Aucun fichier uploade.")
        return

    reference_audio = next(iter(uploaded))
    print(f"Fichier de reference : {reference_audio}")

    text = input("\nEntrez le texte a synthetiser : ")
    language = input("Choisissez la langue (fr/en) [fr]: ").lower() or "fr"

    if language not in ["fr", "en"]:
        print("Langue non supportee. Utilisation du francais par defaut.")
        language = "fr"

    output_name = input("Nom du fichier de sortie [output.wav]: ") or "output.wav"

    result = cloner.clone_voice(text, reference_audio, output_name, language)

    if result:
        print("\nVoici l'audio genere :")
        display(result)

        print("\nVoulez-vous telecharger l'audio ? (o/n)")
        if input().lower() == 'o':
            files.download(output_name)

# Lancer l'application
if __name__ == "__main__":
    run_voice_cloner()
```

```python
import os
import json
import torch
import whisper
from moviepy.editor import VideoFileClip
from transformers import GPTNeoForCausalLM, AutoTokenizer
from typing import List, Dict

class VideoAnalyzer:
    def __init__(self):
        """Initialisation des modèles."""
        self.tokenizer = AutoTokenizer.from_pretrained("/content/gpt-neo-1.3B")
        self.tokenizer.pad_token = self.tokenizer.eos_token
        self.model = GPTNeoForCausalLM.from_pretrained("/content/gpt-neo-1.3B")
        self.model.config.pad_token_id = self.tokenizer.eos_token_id
        self.whisper_model = whisper.load_model("base")
        self.max_tokens = 1024  # Limite de tokens pour GPT-Neo

    def transcribe_video(self, video_path: str) -> List[Dict]:
        """Transcrit la vidéo avec Whisper."""
        print("Transcription de la vidéo en cours...")
        result = self.whisper_model.transcribe(video_path, task="transcribe")
        return result['segments']

    def create_batch_from_segments(self, segments: List[Dict], start_idx: int) -> tuple:
        """Crée un lot de segments qui respecte la limite de tokens."""
        batch_segments = []
        current_text = "Analyze this transcript segment and identify the most important parts to keep for short videos. Focus on technology, business news, announcements, or interesting discussions.\n\nTranscript:\n"

        i = start_idx
        while i < len(segments):
            segment = segments[i]
            new_text = f"[{segment['start']:.2f}s - {segment['end']:.2f}s]: {segment['text']}\n"

            # Vérifier la longueur potentielle
            test_tokens = self.tokenizer.tokenize(current_text + new_text)
            if len(test_tokens) > self.max_tokens and batch_segments:
                break

            current_text += new_text
            batch_segments.append(segment)
            i += 1

        return batch_segments, i, current_text

    def analyze_transcript_with_gpt(self, segments: List[Dict]) -> List[Dict]:
        """Analyse la transcription avec GPT-Neo par lots."""
        all_important_segments = []
        start_idx = 0

        while start_idx < len(segments):
            # Créer un lot de segments
            batch_segments, new_start_idx, prompt = self.create_batch_from_segments(segments, start_idx)

            if not batch_segments:
                break

            # Analyser le lot
            inputs = self.tokenizer(prompt, return_tensors="pt", truncation=True, max_length=self.max_tokens)

            with torch.no_grad():
                outputs = self.model.generate(
                    inputs.input_ids,
                    max_new_tokens=500,
                    num_beams=2,
                    temperature=0.7,
                    do_sample=True,
                    no_repeat_ngram_size=2,
                    pad_token_id=self.tokenizer.eos_token_id
                )

            analysis = self.tokenizer.decode(outputs[0], skip_special_tokens=True)

            # Identifier les segments importants dans ce lot
            current_segment = {"start": None, "end": None, "text": ""}

            for segment in batch_segments:
                if segment['text'] in analysis:
                    if current_segment["start"] is None:
                        current_segment["start"] = segment['start']
                    current_segment["end"] = segment['end']
                    current_segment["text"] += " " + segment['text']

                    # Ajouter le segment s'il atteint la durée minimale
                    if current_segment["end"] - current_segment["start"] >= 60:
                        all_important_segments.append(current_segment.copy())
                        current_segment = {"start": None, "end": None, "text": ""}

            # Ajouter le dernier segment du lot s'il est valide
            if current_segment["start"] is not None and (current_segment["end"] - current_segment["start"]) >= 60:
                all_important_segments.append(current_segment)

            start_idx = new_start_idx
            print(f"Traitement du lot terminé : {start_idx}/{len(segments)} segments")

        return all_important_segments

    def create_short_videos(self, video_path: str, segments: List[Dict], output_folder: str):
        """Crée les shorts vidéo à partir des segments sélectionnés."""
        print("\nCréation des shorts...")
        clip = VideoFileClip(video_path)

        for idx, segment in enumerate(segments, 1):
            start_time, end_time = segment['start'], segment['end']
            duration = end_time - start_time

            print(f"\nShort {idx}: {start_time:.2f}s à {end_time:.2f}s (Durée: {duration:.2f}s)")
            print(f"Contenu: {segment['text'][:200]}...")

            subclip = clip.subclip(start_time, end_time)
            portrait_clip = subclip.resize(newsize=(1080, 1920))

            output_path = os.path.join(output_folder, f"short_{idx:02d}_{start_time:.0f}_{end_time:.0f}.mp4")
            portrait_clip.write_videofile(output_path, codec="libx264", audio_codec="aac", verbose=False)

        clip.close()

def main():
    input_video = "/content/video.mp4"
    output_folder = "/content/shorts_output"
    os.makedirs(output_folder, exist_ok=True)

    try:
        analyzer = VideoAnalyzer()
        segments = analyzer.transcribe_video(input_video)

        with open(os.path.join(output_folder, "transcription.json"), 'w', encoding='utf-8') as f:
            json.dump(segments, f, ensure_ascii=False, indent=2)

        important_segments = analyzer.analyze_transcript_with_gpt(segments)

        with open(os.path.join(output_folder, "important_segments.json"), 'w', encoding='utf-8') as f:
            json.dump({
                "total_segments": len(important_segments),
                "segments": [
                    {
                        "start": seg['start'],
                        "end": seg['end'],
                        "duration": seg['end'] - seg['start'],
                        "text": seg['text']
                    } for seg in important_segments
                ]
            }, f, ensure_ascii=False, indent=2)

        analyzer.create_short_videos(input_video, important_segments, output_folder)

        print("\nTraitement terminé avec succès!")
        print(f"Nombre de shorts créés: {len(important_segments)}")

    except Exception as e:
        print(f"Erreur lors du traitement : {str(e)}")

if __name__ == "__main__":
    main()
```

```python
import os
from transformers import GPTNeoForCausalLM, GPT2Tokenizer, GPTNeoConfig
from moviepy.editor import VideoFileClip, CompositeVideoClip, vfx
from moviepy.config import change_settings
from PIL import Image
import yt_dlp
import re
import logging
import numpy as np
import torch
import librosa
import shutil
import warnings
import json
import ast
from dotenv import load_dotenv
import vosk
import wave

# Load environment variables
load_dotenv()

# Configure ImageMagick for MoviePy
IMAGEMAGICK_BINARY = os.getenv('IMAGEMAGICK_BINARY', 'convert')
change_settings({"IMAGEMAGICK_BINARY": IMAGEMAGICK_BINARY})
Image.ANTIALIAS = Image.Resampling.LANCZOS

logging.basicConfig(level=logging.DEBUG)
torch.set_grad_enabled(False)

warnings.filterwarnings(
    "ignore",
    category=FutureWarning,
    message="You are using `torch.load` with `weights_only=False` .*",
)

def initialize_gpt_neo(model_name="EleutherAI/gpt-neo-1.3B"):
    tokenizer = GPT2Tokenizer.from_pretrained(model_name)
    config = GPTNeoConfig.from_pretrained(model_name)
    model = GPTNeoForCausalLM.from_pretrained(model_name, config=config, torch_dtype=torch.float32)
    logging.info(f"Model {model_name} loaded successfully.")
    return model, tokenizer

def is_youtube_link(input_string):
    youtube_regex = r'(https?://)?(www\.)?(youtube|youtu|youtube-nocookie)\.(com|be)/'
    return re.match(youtube_regex, input_string) is not None

def download_youtube_video(url, output_path):
    try:
        ydl_opts = {
            'format': 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]/best',
            'outtmpl': os.path.join(output_path, 'temp_video.%(ext)s'),
            'verbose': True
        }
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(url, download=True)
            return ydl.prepare_filename(info)
    except Exception as e:
        logging.exception("Error downloading YouTube video")
        return None

def smart_crop(clip, width, height):
    def crop_frame(get_frame, t):
        frame = get_frame(t)
        frame_image = Image.fromarray(frame.astype('uint8'), 'RGB')
        frame_width, frame_height = frame_image.size
        center_x, center_y = frame_width / 2, frame_height / 2
        left, top = max(0, int(center_x - width / 2)), max(0, int(center_y - height / 2))
        cropped_image = frame_image.crop((left, top, left + width, top + height))
        resized_frame = cropped_image.resize((width, height))
        return np.array(resized_frame)
    return clip.fl(crop_frame)

def extract_transcript(audio_path):
    model_path = "model"  # Path to the Vosk model directory
    sample_rate = 16000
    audio, rate = librosa.load(audio_path, sr=sample_rate)

    if rate != sample_rate:
        audio = librosa.resample(audio, rate, sample_rate)

    with wave.open(audio_path, "wb") as wf:
        wf.setnchannels(1)
        wf.setsampwidth(2)
        wf.setframerate(sample_rate)
        wf.writeframes((audio * 32767).astype(np.int16).tobytes())

    model = vosk.Model(model_path)
    rec = vosk.KaldiRecognizer(model, sample_rate)
    rec.SetWords(True)

    transcript = []
    with wave.open(audio_path, "rb") as wf:
        while True:
            data = wf.readframes(4000)
            if len(data) == 0:
                break
            if rec.AcceptWaveform(data):
                result = rec.Result()
                result_dict = json.loads(result)
                if 'result' in result_dict:
                    words = result_dict['result']
                    for word in words:
                        transcript.append({
                            "text": word['word'],
                            "start": word['start'],
                            "end": word['end']
                        })

    final_result = rec.FinalResult()
    final_result_dict = json.loads(final_result)
    if 'result' in final_result_dict:
        words = final_result_dict['result']
        for word in words:
            transcript.append({
                "text": word['word'],
                "start": word['start'],
                "end": word['end']
            })

    return transcript

def time_to_seconds(time_str):
    minutes, seconds = map(int, time_str.split(':'))
    return minutes * 60 + seconds

def analyze_transcript(model, tokenizer, transcript, num_shorts, max_duration=60):
    full_text = " ".join([segment['text'] for segment in transcript])

    prompt = f"""Analyze the following transcript and identify the {num_shorts} most
    interesting segments for short videos. For each segment, provide a brief description
    and start/end times in 'MM:SS' format. Respond with a Python list of dictionaries."""

    inputs = tokenizer(prompt + full_text, return_tensors="pt", truncation=True)

    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_length=512,
            pad_token_id=tokenizer.eos_token_id,
            temperature=0.7,
            num_beams=4
        )

    response = tokenizer.decode(output[0], skip_special_tokens=True)

    try:
        segments = ast.literal_eval(response)
        valid_segments = []
        for segment in segments:
            segment['start'] = time_to_seconds(segment['start_time'])
            segment['end'] = time_to_seconds(segment['end_time'])
            segment['duration'] = segment['end'] - segment['start']
            if segment['duration'] <= max_duration:
                valid_segments.append(segment)
        return valid_segments
    except Exception as e:
        logging.exception(f"Error analyzing response: {e}")
        return []

def create_shorts(input_path, output_path, num_shorts=3, max_duration=60, target_width=1080, target_height=1920):
    model, tokenizer = initialize_gpt_neo()
    is_youtube = is_youtube_link(input_path)
    temp_dir = None
    created_shorts = []

    try:
        if is_youtube:
            temp_dir = os.path.join(os.path.dirname(output_path), "temp_youtube")
            os.makedirs(temp_dir, exist_ok=True)
            input_path = download_youtube_video(input_path, temp_dir)

        with VideoFileClip(input_path) as clip:
            transcript = extract_transcript(input_path)
            segments = analyze_transcript(model, tokenizer, transcript, num_shorts, max_duration)

            for i, segment in enumerate(segments):
                start, end = segment['start'], segment['end']
                with clip.subclip(start, end) as subclip:
                    subclip = smart_crop(subclip, target_width, target_height)
                    output_file = f"{output_path}_part{i + 1}.mp4"
                    subclip.write_videofile(output_file, codec="libx264", audio_codec="aac")
                    created_shorts.append(output_file)

        return created_shorts
    finally:
        if temp_dir:
            shutil.rmtree(temp_dir, ignore_errors=True)

if __name__ == "__main__":
    source_type = input("Enter 'youtube' or 'local': ").strip().lower()
    if source_type == 'youtube':
        url = input("Enter YouTube URL: ")
        create_shorts(url, "output_youtube_short")
    elif source_type == 'local':
        path = input("Enter video path: ")
        create_shorts(path, "output_local_short")
```
